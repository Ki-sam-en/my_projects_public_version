```python
import requests
from requests.exceptions import HTTPError
import json
import locale
```

```python
url = 'http://publication.pravo.gov.ru/api/Documents?Block&Category&SignatoryAuthorityId&DocumentDateFrom=28.01.2024&NameNumberSearchType&Number'
try: 
    response = requests.get(url)
    if response.status_code == 200:
        print('Got it')
except HTTPError as http_err:
    print(f'HTTP error occured: {http_err}')
except Exception as err:
    print(f'Other error occured: {err}')
```

### Параметры ссылки
-----
#### Ссылка на сайт: 
##### 'http://publication.pravo.gov.ru'

#### Сами параметры, которые я брала(есть и другие), чтобы получить ответ(все они необязательные):
##### key - description
##### Block - Код блока публикации
##### Category - Код категории принимающего органа
##### SignatoryAuthorityId - GUID Принявшего органа
##### NameNumberSearchType - Режим поиска по номеру НПА,  возможны следующие значения: 0 – точно, 1 – начинается с, 2 – заканчивается на, 3 – содержит
##### Number - Номер документа (НПА) – режим поиска, в соответствии с параметром «NumberSearchType»
##### DocumentDateFrom - типо от какой даты смотреть документы

##### Дату указала в формате дд.мм.гггг
